```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
FILEPATH_ALL = "all_data.csv"
df = pd.read_csv(FILEPATH_ALL)
ID_TYPE = 'id_applicant'
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
def clientInfo(df, id):
    return df[df['id_applicant'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')

def partnerInfo(df, id):
    return df[df['id_partner'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')
```

```python
df_hired = df[(df['stage_name'] == 'Оформление') & (df['activity_name'] == 'Узнать результаты приема') & (df['result_name'] == 'Принят')]
df_hired = df_hired.reset_index(drop=True)
hired_ids = df_hired[ID_TYPE]
df_filtered = df[df[ID_TYPE].isin(hired_ids)]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
df_filtered.drop(['id_log', 'position_name', 'is_open_action'], axis=1, inplace=True)
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
last = groups.nth(-1)
not_hired = last[last['result_name'] != 'Принят']
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(list(not_hired.index)) == False]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
first = groups.nth(0)
not_created = first[(first['stage_name'] != 'Создание кандидата') & (first['activity_name'] != 'Провести первичную обработку')]
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(not_created.index) == False]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
df_new = pd.DataFrame()
df_new[ID_TYPE] = list(groups.groups)
df_new['first'] = groups.first()['action_date'].values
df_new['last'] = groups.last()['action_date'].values
df_new['first'] = pd.to_datetime(df_new['first'], format="%Y-%m-%d %H:%M:%S")
df_new['last'] = pd.to_datetime(df_new['last'], format="%Y-%m-%d %H:%M:%S")
df_new['diff'] = df_new['last'] - df_new['first']
df_new
```

```python
df_new['diff_days'] = df_new['diff'].apply(lambda x: x.days)
df_new['diff_hours'] = df_new['diff'].apply(lambda x: x.total_seconds()//3600)
df_new['work_days'] = df_new.apply(lambda x: np.busday_count(x['first'].strftime('%Y-%m-%d'), x['last'].strftime('%Y-%m-%d')), axis=1)
df_new
```

```python
df_new.loc[(df_new['work_days'] == 0), 'work_days'] = 1
df_new = df_new[df_new['diff_hours'] >= 24]
df_new
```

```python
quantile = df_new['work_days'].quantile(0.99)
df_new = df_new[df_new['work_days'] < quantile]
df_new
```

```python
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(df_new['id_applicant'])]
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
```

```python
first_action = []
first_valid = []
ids = []
counts = []
for group_name, df_group in groups:
    count = 0
    first = None
    valid = None
    for row_index, row in df_group.iterrows():
        stage = row['stage_name']
        activity = row['activity_name']
        result = row['result_name']
        if stage == 'Интервью с руководителем' and activity == 'Назначить интервью с руководителем':
            if first is None:
                first = row['action_date']
                valid = row['action_date']
            if  result == 'Недозвон' or result == 'Перенос звонка' or result == 'Перенос кандидата' or result == 'Ожидание' or result == 'Нет обратной связи':
                count = count + 1
            else:
                valid = row['action_date']
                break
        
    if not first is None:
        first_action.append(first)
        first_valid.append(valid)
        counts.append(count)
        ids.append(group_name)
        
```

```python
len(ids)
```

```python
df_calls = pd.DataFrame()
df_calls[ID_TYPE] = ids
df_calls['problem_calls'] = counts
df_calls['first_call'] = first_action
df_calls['first_call'] = pd.to_datetime(df_calls['first_call'], format="%Y-%m-%d %H:%M:%S")
df_calls['valid_call'] = first_valid
df_calls['valid_call'] = pd.to_datetime(df_calls['valid_call'], format="%Y-%m-%d %H:%M:%S")
df_calls
```

```python
df_calls['problem_calls'].describe()
```

```python
df_merged = pd.merge(df_new, df_calls, on=ID_TYPE, how='inner')
df_merged
```

```python
df_merged['call_diff'] = df_merged['valid_call'] - df_merged['first_call']
df_merged
```

```python
df_merged['call_diff_days'] = df_merged['call_diff'].apply(lambda x: x.days)
df_merged['call_diff_hours'] = df_merged['call_diff'].apply(lambda x: x.total_seconds()//3600)
df_merged['call_work_days'] = df_merged.apply(lambda x: np.busday_count(x['first_call'].strftime('%Y-%m-%d'), x['valid_call'].strftime('%Y-%m-%d')), axis=1)
df_merged
```

```python
df_merged['call_part'] = df_merged['call_diff'] / df_merged['diff']
df_merged
```

```python
df_final = df_merged[['id_applicant', 'problem_calls', 'call_diff', 'call_diff_days', 'call_diff_hours', 'call_work_days', 'call_part']]
df_final
```
