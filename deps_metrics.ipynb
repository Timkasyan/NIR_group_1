```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
FILEPATH_ALL = "all_data.csv"
df = pd.read_csv(FILEPATH_ALL)
ID_TYPE = 'id_applicant'
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
def clientInfo(df, id):
    return df[df['id_applicant'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')

def partnerInfo(df, id):
    return df[df['id_partner'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')
```

```python
df_hired = df[(df['stage_name'] == 'Оформление') & (df['activity_name'] == 'Узнать результаты приема') & (df['result_name'] == 'Принят')]
df_hired = df_hired.reset_index(drop=True)
df_hired
```

```python
hired_ids = df_hired[ID_TYPE]
df_filtered = df[df[ID_TYPE].isin(hired_ids)]
```

```python
df_hired = df[(df['stage_name'] == 'Оформление') & (df['activity_name'] == 'Узнать результаты приема') & (df['result_name'] == 'Принят')]
df_hired = df_hired.reset_index(drop=True)
df_hired
```

```python
hired_ids = df_hired[ID_TYPE]
df_filtered = df[df[ID_TYPE].isin(hired_ids)]
```

```python
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
df_filtered.drop(['id_log', 'position_name', 'is_open_action'], axis=1, inplace=True)
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
last = groups.nth(-1)
last
```

```python
not_hired = last[last['result_name'] != 'Принят']
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(list(not_hired.index)) == False]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
first = groups.nth(0)
first
```

```python
not_created = first[(first['stage_name'] != 'Создание кандидата') & (first['activity_name'] != 'Провести первичную обработку')]
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(not_created.index) == False]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
df_new = pd.DataFrame()
df_new[ID_TYPE] = list(groups.groups)
df_new['department_id'] = groups.first()['department_id'].values
df_new['actions'] = groups['action_date'].count().values
df_new['first'] = groups.first()['action_date'].values
df_new['last'] = groups.last()['action_date'].values
df_new['first'] = pd.to_datetime(df_new['first'], format="%Y-%m-%d %H:%M:%S")
df_new['last'] = pd.to_datetime(df_new['last'], format="%Y-%m-%d %H:%M:%S")
df_new['diff'] = df_new['last'] - df_new['first']
df_new
```

```python
df_new['diff_days'] = df_new['diff'].apply(lambda x: x.days)
df_new['diff_hours'] = df_new['diff'].apply(lambda x: x.total_seconds()//3600)
```

```python
df_new
```

```python
df_new['work_days'] = df_new.apply(lambda x: np.busday_count(x['first'].strftime('%Y-%m-%d'), x['last'].strftime('%Y-%m-%d')), axis=1)
```

```python
df_new.loc[(df_new['work_days'] == 0), 'work_days'] = 1
```

```python
df_new
```

```python
groups = df_new.groupby('department_id')
df_deps = pd.DataFrame()
df_deps['department_id'] = list(groups.groups)
df_deps['mean_days'] = groups['diff_days'].mean().values
df_deps['median_days'] = groups['diff_days'].median().values
df_deps['mean_wdays'] = groups['work_days'].mean().values
df_deps['median_wdays'] = groups['work_days'].median().values
# df_deps['last'] = groups.last()['action_date'].values
# df_deps['first'] = pd.to_datetime(df_new['first'], format="%Y-%m-%d %H:%M:%S")
# df_deps['last'] = pd.to_datetime(df_new['last'], format="%Y-%m-%d %H:%M:%S")
# df_deps['diff'] = df_new['last'] - df_new['first']
# groups['diff_days'].mean()
```

```python
df_deps
```

```python
df_lowtime = df_new[df_new['diff_hours'] < 24]
df_lowtime
```

```python
groups = df_lowtime.groupby('department_id')
df_deps_low = pd.DataFrame()
df_deps_low['department_id'] = list(groups.groups)
df_deps_low['bad_timing'] = groups['id_applicant'].count().values
df_deps_low
```

```python
df_deps_merged = pd.merge(df_deps, df_deps_low, on='department_id', how='outer')
df_deps_merged['bad_timing'] = df_deps_merged['bad_timing'].fillna(0)
df_deps_merged
```

```python
groups = df_new.groupby('department_id')
df_counts = pd.DataFrame()
df_counts['department_id'] = list(groups.groups)
df_counts['employed'] = groups['id_applicant'].count().values
df_counts
```

```python
df_deps_merged = pd.merge(df_deps_merged, df_counts, on='department_id', how='outer')
df_deps_merged['employed'] = df_deps_merged['employed'].fillna(0)
df_deps_merged
```

```python
df_new_filtered = df_new[df_new['diff_hours'] >= 24]
df_new_filtered
```

```python
quantile = df_new_filtered['work_days'].quantile(0.99)
df_new_filtered = df_new_filtered[df_new_filtered['work_days'] < quantile]
df_new_filtered
```

```python
groups = df_new_filtered.groupby('department_id')
df_deps_filtered = pd.DataFrame()
df_deps_filtered['department_id'] = list(groups.groups)
df_deps_filtered['mean_days_f'] = groups['diff_days'].mean().values
df_deps_filtered['median_days_f'] = groups['diff_days'].median().values
df_deps_filtered['mean_wdays_f'] = groups['work_days'].mean().values
df_deps_filtered['median_wdays_f'] = groups['work_days'].median().values
```
