```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
FILEPATH_ALL = "all_data.csv"
df = pd.read_csv(FILEPATH_ALL)
ID_TYPE = 'id_applicant'
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
def clientInfo(df, id):
    return df[df['id_applicant'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')

def partnerInfo(df, id):
    return df[df['id_partner'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')
```

```python
df_calls = pd.read_csv('call_metrics.csv')
df_deps = pd.read_csv('deps_metrics.csv')
df_fb = pd.read_csv('fb_metrics.csv')
```

```python
df_calls = df_calls.drop('Unnamed: 0', axis=1)
df_calls
```

```python
df_deps = df_deps.drop('Unnamed: 0', axis=1)
df_deps
```

```python
df_fb = df_fb.drop('Unnamed: 0', axis=1)
df_fb
```

```python
df_main = pd.DataFrame()
df_main['id_applicant'] = df_calls['id_applicant']
df_main['problem_calls'] = df_calls['problem_calls']
df_main
```

```python
df_main = pd.merge(df_main, df_fb[['id_applicant', 'first', 'fb_work_days']], on='id_applicant', how='inner')
```

```python
df_main
```

```python
df_tmp = df[df[ID_TYPE].isin(df_main[ID_TYPE])]
groups = df_tmp.groupby(ID_TYPE)
first = groups.nth(0)
```

```python
df_tmp_deps = pd.DataFrame(first['department_id']).reset_index()
df_tmp_deps
```

```python
df_main = pd.merge(df_main, df_tmp_deps, on='id_applicant', how='inner')
df_main
```

```python
df_main = df_main[['id_applicant', 'department_id', 'fb_work_days', 'problem_calls', 'first']]
df_main
```

```python
df_main = pd.merge(df_main, df_deps[['department_id', 'dep_tth_diff']], on='department_id', how='inner')
df_main
```

```python
df_main['first'] = pd.to_datetime(df_main['first'], format="%Y-%m-%d %H:%M:%S")
df_main['month'] = df_main['first'].apply(lambda x: x.month)
```

```python
df_main['season_1'] = df_main['month'].apply(lambda x: 1 if (x < 3 or x == 12) else 0)
df_main['season_2'] = df_main['month'].apply(lambda x: 1 if (x >= 3 and x <= 5) else 0)
df_main['season_3'] = df_main['month'].apply(lambda x: 1 if (x >= 6 and x <= 8) else 0)
```

```python
df_main
```

```python
df_final = df_main.drop(['first'], axis=1)
df_final
```

```python
df_final.to_csv('regression_data_v1.csv')
```
