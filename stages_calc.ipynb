```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
FILEPATH_ALL = "all_data.csv"
df = pd.read_csv(FILEPATH_ALL)
ID_TYPE = 'id_applicant'
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
def clientInfo(df, id):
    return df[df['id_applicant'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')

def partnerInfo(df, id):
    return df[df['id_partner'] == id].drop(['is_open_action', 'position_name'], axis=1).sort_values('action_date')
```

```python
clientInfo(df, 2196746)
```

```python
df_hired = df[(df['stage_name'] == 'Оформление') & (df['activity_name'] == 'Узнать результаты приема') & (df['result_name'] == 'Принят')]
df_hired = df_hired.reset_index(drop=True)
hired_ids = df_hired[ID_TYPE]
df_filtered = df[df[ID_TYPE].isin(hired_ids)]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
df_filtered.drop(['id_log', 'position_name', 'is_open_action'], axis=1, inplace=True)
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
last = groups.nth(-1)
not_hired = last[last['result_name'] != 'Принят']
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(list(not_hired.index)) == False]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
first = groups.nth(0)
not_created = first[(first['stage_name'] != 'Создание кандидата') & (first['activity_name'] != 'Провести первичную обработку')]
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(not_created.index) == False]
df_filtered = df_filtered.sort_values([ID_TYPE,'action_date'])
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
df_new = pd.DataFrame()
df_new[ID_TYPE] = list(groups.groups)
df_new['first'] = groups.first()['action_date'].values
df_new['last'] = groups.last()['action_date'].values
df_new['first'] = pd.to_datetime(df_new['first'], format="%Y-%m-%d %H:%M:%S")
df_new['last'] = pd.to_datetime(df_new['last'], format="%Y-%m-%d %H:%M:%S")
df_new['diff'] = df_new['last'] - df_new['first']
df_new
```

```python
df_new['diff_days'] = df_new['diff'].apply(lambda x: x.days)
df_new['diff_hours'] = df_new['diff'].apply(lambda x: x.total_seconds()//3600)
df_new['work_days'] = df_new.apply(lambda x: np.busday_count(x['first'].strftime('%Y-%m-%d'), x['last'].strftime('%Y-%m-%d')), axis=1)
df_new
```

```python
df_new.loc[(df_new['work_days'] == 0), 'work_days'] = 1
df_new = df_new[df_new['diff_hours'] >= 24]
df_new
```

```python
quantile = df_new['work_days'].quantile(0.99)
df_new = df_new[df_new['work_days'] < quantile]
df_new
```

```python
df_filtered = df_filtered[df_filtered[ID_TYPE].isin(df_new['id_applicant'])]
df_filtered
```

```python
groups = df_filtered.groupby(ID_TYPE)
```

```python
stage1_end = []
stage2_start = []
stage2_end = []
stage3_start = []
stage3_end = []
stage4_start = []
ids = []
for group_name, df_group in groups:
    st1_end = None
    st2_start = None
    st2_end = None
    st3_start = None
    st3_end = None
    st4_start = None
    prev_date = None
    for row_index, row in df_group.iterrows():
        stage = row['stage_name']
        activity = row['activity_name']
        result = row['result_name']
        if stage == 'Интервью с руководителем' and (st2_start is None):
            st2_start = row['action_date']
            st1_end = prev_date
        if stage == 'СБ' and (st2_start is not None and st3_start is None):
            st3_start = row['action_date']
            st2_end = prev_date
        if (stage == 'Оформление' or stage == 'Назначение времени на оформление/медосмотр') and (st2_start is not None and st3_start is not None and st4_start is None):
            st4_start = row['action_date']
            st3_end = prev_date
            break
        prev_date = row['action_date']
        
    if (not st1_end is None) and (not st2_start is None) and (not st2_end is None) and (not st3_start is None) and (not st3_end is None) and (not st4_start is None):
        stage1_end.append(st1_end)
        stage2_start.append(st2_start)
        stage2_end.append(st2_end)
        stage3_start.append(st3_start)
        stage3_end.append(st3_end)
        stage4_start.append(st4_start)
        ids.append(group_name)
        
```

```python
len(ids)
```

```python
df_stages = pd.DataFrame()
df_stages[ID_TYPE] = ids
df_stages['stage1_end'] = stage1_end
df_stages['stage1_end'] = pd.to_datetime(df_stages['stage1_end'], format="%Y-%m-%d %H:%M:%S")
df_stages['stage2_start'] = stage2_start
df_stages['stage2_start'] = pd.to_datetime(df_stages['stage2_start'], format="%Y-%m-%d %H:%M:%S")
df_stages['stage2_end'] = stage2_end
df_stages['stage2_end'] = pd.to_datetime(df_stages['stage2_end'], format="%Y-%m-%d %H:%M:%S")
df_stages['stage3_start'] = stage3_start
df_stages['stage3_start'] = pd.to_datetime(df_stages['stage3_start'], format="%Y-%m-%d %H:%M:%S")
df_stages['stage3_end'] = stage3_end
df_stages['stage3_end'] = pd.to_datetime(df_stages['stage3_end'], format="%Y-%m-%d %H:%M:%S")
df_stages['stage4_start'] = stage4_start
df_stages['stage4_start'] = pd.to_datetime(df_stages['stage4_start'], format="%Y-%m-%d %H:%M:%S")
df_stages
```

```python
df_merged = pd.merge(df_new, df_stages, on=ID_TYPE, how='inner')
df_merged
```

```python
df_final = df_merged[['id_applicant','first','stage1_end','stage2_start','stage2_end','stage3_start','stage3_end','stage4_start','last']]
df_final
```

```python
df_final.to_csv('stages.csv')
```
