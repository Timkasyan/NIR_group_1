```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse


from sklearn.linear_model import LinearRegression
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('regression_stages.csv')
ID_TYPE = 'id_applicant'
df= df.drop('Unnamed: 0', axis=1)
df
```

```python
df = df.drop(['id_applicant','1_start','1_end','2_start','2_end','3_start','3_end','4_start','4_end',],axis=1)
df
```

```python
from statsmodels.tools.eval_measures import rmse
def calc_error(y, y_pred):
    diff = y_pred - y
    diff = diff.apply(lambda x: abs(x))
    mean = diff.mean()
    median = diff.median()
    rmse_err = rmse(y, y_pred)
    print(f'mean err: {mean}')
    print(f'median err: {median}')
    print(f'rmse: {rmse_err}')
```

```python
# 2 3 4 const
X = df.drop(['1_full', '2_work','3_work','4_work','2_delay','3_delay','4_delay','tth'], axis=1)
y = df['tth']

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
calc_error(y, est2.predict(X2))
```

```python
# 2 3 4 no const
X = df.drop(['1_full', '2_work','3_work','4_work','2_delay','3_delay','4_delay','tth'], axis=1)
y = df['tth']

est = sm.OLS(y, X)
est2 = est.fit()
print(est2.summary())
calc_error(y, est2.predict(X))
```
