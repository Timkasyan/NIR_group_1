```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
df = pd.read_csv('regression_data_v1.csv')
df_tth = pd.read_csv('tth_metrics.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df_tth = df_tth.drop('Unnamed: 0', axis=1)
df_tth
```

```python
df_tth = df_tth.rename(columns={"id": "id_applicant", "work_days": "tth"})
df_tth
```

```python
df = pd.merge(df, df_tth[['id_applicant', 'first']], on='id_applicant', how='inner')
```

```python
df
```

```python
df['first'] = pd.to_datetime(df['first'], format="%Y-%m-%d %H:%M:%S")
df['year'] = df['first'].apply(lambda x: x.year)
df
```

```python
df.drop('first', axis=1, inplace=True)
df
```

```python
df_deps = pd.read_csv('deps_metrics.csv')
df_deps = df_deps.drop('Unnamed: 0', axis=1)
df_deps
```

```python
df = pd.merge(df, df_deps[['department_id', 'employed']], on='department_id', how='inner')
df
```

```python
df = df[['id_applicant','department_id','year','month','fb_work_days','problem_calls','dep_tth_diff', 'employed','season_1','season_2','season_3', 'tth']]
df
```

```python
df.to_csv('regression_data_v2.csv')
```
