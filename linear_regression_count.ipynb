```python

df_reg_5 = pd.read_csv('regression_data_v5.csv')
ID_TYPE = 'id_applicant'
df_reg_5 = df_reg_5.drop('Unnamed: 0', axis=1)
df_reg_5
```

```python
df_2018 = df_reg_5[df_reg_5['year'] == 2018]
df_2019 = df_reg_5[df_reg_5['year'] == 2019]
df_2020 = df_reg_5[df_reg_5['year'] == 2020]
df_2020
```

```python
def splitMonth(df_reg_5):
    df_reg_5['mon1'] = df_reg_5['month'].apply(lambda x: 1 if (x == 1) else 0)
    df_reg_5['mon2'] = df_reg_5['month'].apply(lambda x: 1 if (x == 2) else 0)
    df_reg_5['mon3'] = df_reg_5['month'].apply(lambda x: 1 if (x == 3) else 0)
    df_reg_5['mon4'] = df_reg_5['month'].apply(lambda x: 1 if (x == 4) else 0)
    df_reg_5['mon5'] = df_reg_5['month'].apply(lambda x: 1 if (x == 5) else 0)
    df_reg_5['mon6'] = df_reg_5['month'].apply(lambda x: 1 if (x == 6) else 0)
    df_reg_5['mon7'] = df_reg_5['month'].apply(lambda x: 1 if (x == 7) else 0)
    df_reg_5['mon8'] = df_reg_5['month'].apply(lambda x: 1 if (x == 8) else 0)
    df_reg_5['mon9'] = df_reg_5['month'].apply(lambda x: 1 if (x == 9) else 0)
    df_reg_5['mon10'] = df_reg_5['month'].apply(lambda x: 1 if (x == 10) else 0)
    df_reg_5['mon11'] = df_reg_5['month'].apply(lambda x: 1 if (x == 11) else 0)
    df_reg_5['mon12'] = df_reg_5['month'].apply(lambda x: 1 if (x == 12) else 0)
    return df_reg_5
```

```python
df_2018 = splitMonth(df_2018)
df_2019 = splitMonth(df_2019)
df_2020 = splitMonth(df_2020)
```

```python
def countLinearRegression(df_reg_5):
    X5 = df_reg_5[['fb_work_days',
                  'problem_calls',
                  'dep_tth_diff',
                  'employed',
#                   'mon1',
#                   'mon2',
#                   'mon3',
#                   'mon4',
#                   'mon5',
#                   'mon6',
#                   'mon7',
#                   'mon8',
#                   'mon9',
#                   'mon10',
#                   'mon11',
#                   'mon12',
               'interview_work_days']]
    Y5 = df_reg_5['tth']
    regr_5 = LinearRegression()
    regr_5.fit(X5, Y5)
    # b0
    print('b0: ',regr_5.intercept_)
    # b1,b2 ... bn
    print('b1,b2 ... bn: ',regr_5.coef_)
    # R^2
    print('R^2: ',regr_5.score(X5,Y5))
```
