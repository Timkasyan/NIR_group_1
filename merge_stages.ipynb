```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
df = pd.read_csv('stages.csv')
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df['1_start'] = pd.to_datetime(df['first'], format="%Y-%m-%d %H:%M:%S")
df['1_end'] = pd.to_datetime(df['stage1_end'], format="%Y-%m-%d %H:%M:%S")
df['2_start'] = pd.to_datetime(df['stage2_start'], format="%Y-%m-%d %H:%M:%S")
df['2_end'] = pd.to_datetime(df['stage2_end'], format="%Y-%m-%d %H:%M:%S")
df['3_start'] = pd.to_datetime(df['stage3_start'], format="%Y-%m-%d %H:%M:%S")
df['3_end'] = pd.to_datetime(df['stage3_end'], format="%Y-%m-%d %H:%M:%S")
df['4_start'] = pd.to_datetime(df['stage4_start'], format="%Y-%m-%d %H:%M:%S")
df['4_end'] = pd.to_datetime(df['last'], format="%Y-%m-%d %H:%M:%S")
df
```

```python
df = df.drop(['first','stage1_end','stage2_start','stage2_end','stage3_start','stage3_end','stage4_start','last'], axis=1)
df
```

```python
df['1_full'] = df['1_end'] - df['1_start']
df['2_full'] = df['2_end'] - df['1_end']
df['3_full'] = df['3_end'] - df['2_end']
df['4_full'] = df['4_end'] - df['3_end']
df['2_work'] = df['2_end'] - df['2_start']
df['3_work'] = df['3_end'] - df['3_start']
df['4_work'] = df['4_end'] - df['4_start']
df['2_delay'] = df['2_start'] - df['1_end']
df['3_delay'] = df['3_start'] - df['2_end']
df['4_delay'] = df['4_start'] - df['3_end']

df
```

```python
df['1_full'] = df['1_full'].apply(lambda x: x.total_seconds()/86400)
df['2_full'] = df['2_full'].apply(lambda x: x.total_seconds()/86400)
df['3_full'] = df['3_full'].apply(lambda x: x.total_seconds()/86400)
df['4_full'] = df['4_full'].apply(lambda x: x.total_seconds()/86400)
df['2_work'] = df['2_work'].apply(lambda x: x.total_seconds()/86400)
df['3_work'] = df['3_work'].apply(lambda x: x.total_seconds()/86400)
df['4_work'] = df['4_work'].apply(lambda x: x.total_seconds()/86400)
df['2_delay'] = df['2_delay'].apply(lambda x: x.total_seconds()/86400)
df['3_delay'] = df['3_delay'].apply(lambda x: x.total_seconds()/86400)
df['4_delay'] = df['4_delay'].apply(lambda x: x.total_seconds()/86400)
df
```

```python
df['tth'] = df['4_end'] - df['1_start']
df['tth'] = df['tth'].apply(lambda x: x.total_seconds()/86400)
df
```
